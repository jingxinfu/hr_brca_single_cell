# Tumor

```python
celltype='Tumor'
MP_gmt_path = f'{Manuscript_RESULT}/MPs/{celltype}/MP_Programs.gmt'
MP_anno_path = f'{Manuscript_RESULT}/MPs/{celltype}/MetaProgram_Annotation.csv'
degs = pd.read_excel(f"{Manuscript_RESULT}/DEGs_Response/{celltype}.xlsx",sheet_name='0-I',index_col=0)
degs = pd.read_excel(f"{Manuscript_RESULT}/DEGs_Response/{celltype}.xlsx",sheet_name='0-I',index_col=0)
gmt = pd.read_table(MP_gmt_path ,header=None,index_col=0).drop([1],axis=1).T
anno = pd.read_csv(MP_anno_path,index_col=0)
MP_anno_map = anno[['MetaProgram','MPs']].drop_duplicates().set_index('MetaProgram')['MPs'].to_dict()
gmt.rename(columns=MP_anno_map,inplace=True)
gmt.head()
```

```python
fig,ax= plot_volcano_df(
    degs,
    x='log2FoldChange',
    y='padj',
    top=20,
    figsize=(10, 5)
)
ax.set(title='NR <-> R');
```

```python
show_genes = degs.index[degs.index.isin(gmt['ER-I']) & (degs.padj<0.05) & (degs.log2FoldChange<.5)].tolist()
fig,ax= plot_volcano_df(
    degs,
    x='log2FoldChange',
    y='padj',
    show_list=show_genes,
    figsize=(5, 5),
    dpi=200
)
ax.set(title='Baseline tumor');
xmax = np.abs(ax.get_xlim()).max()
ax.set_xlim([-xmax,xmax])
center=0.5
name ='Favorable\n(RCB=0-I)'
ax.annotate(name, xy=(center, -0.15), xycoords='axes fraction', xytext=(1.1, -0.15), 
            arrowprops=dict(arrowstyle="<-", color='#006d77'),ha='right',va='center')
name ='Unfavorable\n(RCB=II-III)'
ax.annotate(name, xy=(center, -0.15), xycoords='axes fraction', xytext=(0, -0.15), 
            arrowprops=dict(arrowstyle="<-",  color='#e29578'),ha='right',va='center')
```

```python
show_genes = degs.index[degs.index.isin(gmt['EMT-III']) & (degs.padj<0.05) & (degs.log2FoldChange>.5)].tolist()
fig,ax= plot_volcano_df(
    degs,
    x='log2FoldChange',
    y='padj',
    show_list=show_genes,
    figsize=(5, 5),
    dpi=200
)
ax.set(title='Baseline tumor');
xmax = np.abs(ax.get_xlim()).max()
ax.set_xlim([-xmax,xmax])
center=0.5
name ='Favorable\n(RCB=0-I)'
ax.annotate(name, xy=(center, -0.15), xycoords='axes fraction', xytext=(1.1, -0.15), 
            arrowprops=dict(arrowstyle="<-", color='#006d77'),ha='right',va='center')
name ='Unfavorable\n(RCB=II-III)'
ax.annotate(name, xy=(center, -0.15), xycoords='axes fraction', xytext=(0, -0.15), 
            arrowprops=dict(arrowstyle="<-",  color='#e29578'),ha='right',va='center')
```

# Macrophage

```python
celltype='Macs'
MP_gmt_path = f'{Manuscript_RESULT}/MPs/{celltype}/MP_Programs.gmt'
MP_anno_path = f'{Manuscript_RESULT}/MPs/{celltype}/MetaProgram_Annotation.csv'
degs = pd.read_excel(f"{Manuscript_RESULT}/DEGs_Response/{celltype}.xlsx",sheet_name='0-I',index_col=0)
gmt = pd.read_table(MP_gmt_path ,header=None,index_col=0).drop([1],axis=1).T
anno = pd.read_csv(MP_anno_path,index_col=0)
MP_anno_map = anno[['MetaProgram','MPs']].drop_duplicates().set_index('MetaProgram')['MPs'].to_dict()
gmt.rename(columns=MP_anno_map,inplace=True)
gmt.head()
```

```python
fig,ax= plot_volcano_df(
    degs,
    x='log2FoldChange',
    y='padj',
    top=30,
    figsize=(10, 10)
)
ax.set(title='Macrophage')
```

```python
for c in gmt:
    show_genes = degs.index[degs.index.isin(gmt[c]) & (degs.padj<0.05) & (degs.log2FoldChange.abs()>.5)].tolist()
    if len(show_genes) ==0:
        continue
    fig,ax= plot_volcano_df(
        degs,
        x='log2FoldChange',
        y='padj',
        show_list=show_genes,
        figsize=(10, 5)
    )
    ax.set(title=f'NR <-> R\n{c}');
```
