# CCLE Breast Cell line signature enrichment

```python
# load CCLE cell line annotation
annot_ccle = pd.read_csv(
    f'{EXTERNAL_DATA}/CCLE/Hotspot_Mutations_subsetted.csv')[['cell_line_display_name','lineage_5']].dropna()
annot_ccle['lineage_5'] = annot_ccle['lineage_5'].str.replace('pos','+').str.replace('neg','-')
annot_ccle['lineage_5'] = annot_ccle['lineage_5'].str.replace('ER- HER2-','TNBC')
annot_ccle = annot_ccle.set_index('cell_line_display_name')['lineage_5'].to_dict()
# add the lineage subtype to cell line
ccle_breast_sets ={ k:v for k,v in gp.get_library('CCLE_Proteomics_2020').items() if 'BREAST' in k}
ccle_breast_sets_rename = {k.split(' ')[0] +' ' +annot_ccle[k.split(' ')[0]]:v for k,v in ccle_breast_sets.items() if k.split(' ')[0] in annot_ccle}
```

## EMT-III signature is enriched in TNBC cell line

```python
deg = degs_dict['Tumor.EMT-III']
pre_res = gp.prerank(rnk=deg.loc[deg.padj<0.05,'stat'],
                     gene_sets=ccle_breast_sets_rename,
                     min_size=50,
                     max_size=1000,
                     permutation_num=500, # reduce number to speed up testing
                     outdir=None, # don't write to disk
                     seed=6,
                     verbose=True, # see what's going on behind the scenes
                    )
terms = pre_res.res2d.loc[pre_res.res2d['FDR q-val']<0.05,:].sort_values('NES',ascending=False).head(3)['Term'].tolist()
axs = pre_res.plot(terms=terms,show_ranking=False,figsize=(15,10))
```

## ER-I signature is enriched in ER+ cell line

```python
deg = degs_dict['Tumor.ER-I']
pre_res = gp.prerank(rnk=deg.loc[deg.padj<0.05,'stat'],
                     gene_sets=ccle_breast_sets_rename,
                     min_size=50,
                     max_size=1000,
                     permutation_num=500, # reduce number to speed up testing
                     outdir=None, # don't write to disk
                     seed=6,
                     verbose=True, # see what's going on behind the scenes
                    )
terms = pre_res.res2d.loc[pre_res.res2d['FDR q-val']<0.05,:].sort_values('NES',ascending=False).head(3)['Term'].tolist()
axs = pre_res.plot(terms=terms,show_ranking=False,figsize=(15,10))
```
