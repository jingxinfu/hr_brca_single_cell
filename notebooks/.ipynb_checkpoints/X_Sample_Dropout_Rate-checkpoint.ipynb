```python
REPO = '..'
RESULT_TABLE = f'{REPO}/data/table'
EXTERNAL_DATA = f'{REPO}/data/external'
import sys
import commentjson
sys.path.append(REPO)
import numpy as np
import pandas as pd
import warnings
import seaborn as sns
from settings import COLOR_PAlETTE
warnings.filterwarnings("ignore")
%load_ext autoreload
```

# Load data

```python
## CCG sample submission form
ccg = pd.read_excel(f"{EXTERNAL_DATA}/20220919_CCG1112d_MO_SampleSubmissionSheet_sorted_manual.xlsx",skiprows=15).dropna(subset='MO_Batch')
ccg['sample'] = ccg['Sample ID'].map(lambda x:'P'+x.split('_')[1]+'.'+x.split('_')[2])
## samples included in the paper
cohort = pd.read_excel( f"{RESULT_TABLE}/Supplementary Table 1.xlsx")
sample_sheet = pd.concat([cohort.loc[cohort.Tech=="5' v2 gene expression",['sample']],
           ccg[['sample']]]
         ).set_index('sample')
sample_sheet['Library_Construction'] = 'Fail'
sample_sheet.loc[sample_sheet.index.isin(cohort['sample']),'Library_Construction']= 'Success'

## Merge clinial and bulk info 
clinMap = pd.read_csv(f"{EXTERNAL_DATA}/clin_IHC.csv",index_col=0).dropna(subset=['CCG Piece ID'])
clinMap['sample'] = clinMap['CCG Piece ID'].map(lambda x:'P'+'.'.join(x.split('_')[1:3]))
final_sheet = sample_sheet.merge(clinMap,left_index=True,right_on='sample',how='left')
final_sheet['Timepoint'] =final_sheet.Timepoint.map({
    'AfterSurg':'Surgery',
    'Baseline':'Baseline',
    'Surg+AC':'Pre-surgery',
    'W3D1':'W3D1',
    'W7D1':'W7D1'
})
final_sheet['Treatment_Arm'] =final_sheet.Treatment_Arm.map({
    'Chemo->ICI':'Chemo->Combo',
    'ICI->Chemo':'ICI->Combo',
})
final_sheet['BestResponse']=final_sheet['BestResponse'].map({
    '0-I':'favorable response\n(RCB 0-I)',
    'II-III':'unfavorable response\n(RCB II-III)'
})
# final_sheet['Timepoint'] = pd.Categorical(final_sheet['Timepoint'],
#                                    ['Baseline','W3D1','W7D1','Pre-surgery','Surgery'])
```

# Visualize the library construction failure rate

```python
## counting the number of samples that fail/successfully  library in different groups
data = final_sheet.groupby(['Timepoint','BestResponse','Treatment_Arm'])['Library_Construction'].value_counts().rename('N').reset_index()
data['Timepoint'] = pd.Categorical(data['Timepoint'],
                                   ['Baseline','W3D1','W7D1','Pre-surgery','Surgery'])

g = sns.catplot(kind='bar',data=data,x='Timepoint',y='N',
                hue='Library_Construction',col='Treatment_Arm',row='BestResponse',
                palette={'Fail':'salmon','Success':'green'},
                height=3,
                aspect=1.5
           )
g.set_titles("{col_name}\nRCB={row_name}")
g.set(ylabel='Number of Samples')
g.legend.set_title('Library construction')
for ax in g.axes.flatten():
    ax.bar_label(ax.containers[0])
    ax.bar_label(ax.containers[1])

g.savefig('../data/figure/X_Suppl_library_failure_rate.pdf',dpi=150)
```

```python
rate = final_sheet.groupby(['Timepoint','BestResponse'])['Library_Construction'].value_counts(normalize=True).rename('Failure Rate').reset_index()
rate = rate.loc[rate.Library_Construction=='Fail',:]
rate['Timepoint'] = pd.Categorical(rate['Timepoint'],
                                   ['Baseline','W3D1','W7D1','Pre-surgery','Surgery'])

g = sns.catplot(kind='bar',data=rate,x='Timepoint',y='Failure Rate',
               hue='BestResponse',
                height=3,
                aspect=2,
                palette=COLOR_PAlETTE['BestResponse']
           )
for ax in g.axes.flatten():
    ax.spines[['left']].set_visible(False)
    ax.yaxis.set_visible(False)
    ax.bar_label(ax.containers[0],fmt='{:.1%}')
    ax.bar_label(ax.containers[1],fmt='{:.1%}')
g.set(title="Faliure Rate",xlabel='',ylabel='')
```

# Patient and disease characteristics in different timepoint

```python
from tableone import TableOne
import scipy
def my_custom_test(group1, group2):
    """
    Hypothesis test for test_self_defined_statistical_tests
    """
    my_custom_test.__name__ = "Wilcoxon Test "
    _, pval= scipy.stats.ranksums(group1, group2)
    return pval
    
with pd.ExcelWriter('patient_tumor_characteristics.xlsx') as writer:
    data = final_sheet.loc[(final_sheet.Library_Construction!='Fail')&
                                (final_sheet.Timepoint.isin(['Baseline'])),['Timepoint','age','stage','histology','er_status','BestResponse']]
    mytable = TableOne(data,rename={'er_status': 'HR status'},
                 nonnormal=['age'],
        )
    mytable.to_excel(writer,sheet_name='Baseline')
                         
    for timepoint in ['W3D1','W7D1','Pre-surgery','Surgery']:
        data = final_sheet.loc[(final_sheet.Library_Construction!='Fail')&
                                (final_sheet.Timepoint.isin(['Baseline',timepoint])),['Timepoint','age','stage','histology','er_status','BestResponse']]
        mytable = TableOne(data,groupby='Timepoint',
                rename={'er_status': 'HR status'},
                 nonnormal=['age'],
                htest={'age':my_custom_test},        
                 htest_name=True,pval = True
        )
        mytable.to_excel(writer,sheet_name=timepoint)
```
