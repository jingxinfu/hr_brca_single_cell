```python
REPO = '../'
RESULT_TABLE = f'{REPO}/data/table'
RESULT_OBJ = f'{REPO}/data/object'
FIGURE_FOLDER= f'{REPO}/data/figure'
SETTING_FOLDER = f'{REPO}/data/setting'

import numpy as np
import pandas as pd
from comut import comut
import scPipe as sp
import scanpy as sc
import seaborn as sns
import warnings
import sys
import commentjson
sys.path.append(REPO)
from utils.visual import plot_proportion_barplot
import matplotlib.pyplot as plt
from settings import COLOR_PAlETTE
with open(f'{SETTING_FOLDER}/ColorMap.json','r') as f:
    colormap = commentjson.load(f)
COLOR_PAlETTE.update(colormap)
plt.style.use(f'{REPO}/paper.mplstyle')
warnings.filterwarnings("ignore")
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
adata = sc.read(f'{RESULT_OBJ}/gex_all.h5ad')
obs = pd.read_csv(f'{RESULT_TABLE}/GEX_OBS_Lineage.csv',index_col=0)
sample_meta =  pd.read_excel(f'{RESULT_TABLE}/Supplementary Table 1.xlsx',index_col=0).replace(np.nan,'N/A')
obs = obs.reset_index().merge(sample_meta,left_on='Sample',right_on='CCG_ID',how='left').set_index('index')
sc.pp.filter_genes(adata,min_cells=3) # consistent with the clustering method
sp.utils.attach_attr_from_pickle(data=adata,pickle_file=f'{RESULT_OBJ}/leiden_2000variable_genes/embeddings.pickle')
adata = adata[obs.index,:]
for c in ['Sample_Short','Timepoint','Treatment_Arm','Tech','Patient','Compartment']:
    adata.obs[c] = obs[c]
```

```python
with open (f'{RESULT_TABLE}/canonical_breast.json','r') as f:
    marker = commentjson.load(f)
with pd.ExcelWriter(f"{RESULT_TABLE}/Supplementary Dataset 1.xlsx") as writer:  
    adata.obs.loc[:,:'Sample_Short'].to_excel(writer,sheet_name='Cell stats',index=False)
    pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in marker.items() ])).to_excel(writer,sheet_name='Canonical Marker',index=False)
```

## Distribution of cell statistics

```python
fig,axs=plt.subplots(1,2,figsize=(10,10),sharey=True,dpi=150)
ax =axs[0]
sns.violinplot(data=adata.obs,y='Sample_Short',x='n_genes_by_counts',ax=ax)
ax.set(ylabel='',xlabel='# of detected genes')
ax =axs[1]
sns.violinplot(data=adata.obs,y='Sample_Short',x='pct_counts_mito',ax=ax)
ax.set(ylabel='',xlabel='% of mitochondrial reads')
fig.savefig(f'{FIGURE_FOLDER}/GEX_cell_statistics.pdf')
```

### UMAPs

```python
for c in ['Timepoint','Tech','Treatment_Arm','Patient']:
    fig = sc.pl.umap(adata,color=[c],return_fig=True)
    fig.savefig(f'{FIGURE_FOLDER}/UMAP/GEX_AllCell_{c}.png')
```

```python
adata.obs['Compartment'] = pd.Categorical(adata.obs['Compartment'],['Epithelial', 'Immune', 'Stromal'])
adata.uns['Compartment_colors'] = ['#628395','#DBAD6A','#D0CE7C']
for c in ['Timepoint','Tech','Treatment_Arm']:
    g = plot_proportion_barplot(adata,c,'Compartment')
    g.save(f'{FIGURE_FOLDER}/stacked_proportion_plot/GEX_Broad_Celltype_{c}.pdf')
```
